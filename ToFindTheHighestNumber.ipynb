```python
import os
from collections import Counter
import pandas as pd
```

```python
#creating the functions to find the repeated files sorted in descending order
folder_path = "xxxx"
def list_files_in_confirmed_folder(folder_path): 
    return [file for file in os.listdir(folder_path)]
files= list_files_in_confirmed_folder(folder_path)

def extract_client_names(files, folder_path):
    client_names=[]
    for file in files:
        file_name_parts = os.path.splitext(file)[0].split('_')
        # The client name is assumed to be the last part of the file name
        client_name = file_name_parts[-1]
        client_names.append(client_name)
    return client_names
client_names = extract_client_names(files, folder_path)
print("Client names extracted from files:", client_names)
    # Count occurrences of each client name
client_counts = Counter(client_names)
    # Sort client counts by value (count) in ascending order
sorted_clients = sorted(client_counts.items(), key=lambda x: x[1], reverse=True)
print("Client names and their counts in descending order:")
for client, count in sorted_clients:
    print(f"{client}: {count}")

df = pd.DataFrame(sorted_clients, columns=["Client", "Count"])

# Save DataFrame to Excel
# df.to_excel("client_counts.xlsx", index=False)
```
